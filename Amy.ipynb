```python
from pdfminer.high_level import extract_text, extract_pages
from pdfminer.layout import LAParams, LTTextBox, LTTextContainer
import re
```

```python
text_o = extract_text('data/Ayllupi-yachasunchik-1-ñiqi-qullaw-qichwapi.pdf', page_numbers=range(10, 106))
text_o
```

# Limpieza de texto

```python
text_o = text_o.replace('\x0c', '\x0c\n')
lines = text_o.split('\n')
new_lines = [lines[0]] + [line for prev_line, line in zip(lines, lines[1:]) if line != prev_line]
text_o = '\n'.join(new_lines)
```

```python
# get rid of the \n\n and \n in the text. make sure to put a ' ' in place of the \n\n 
text = re.sub(r'\n', ' ', text_o)
text
```

## Willakuykuna-mayt’u-Collao

```python
# Define the patterns for the text to remove
# footer patterns
pattern1 = r'\d*?\s*\x0c'
pattern2 = r'/\s*\w+'
# author patterns
pattern3 = r'(Willakuq:|Qillqaq:|Willaquq:|Kastilla simipi qillqaq:|Qichwaman tikraspa qillqaq:|Chayllam chay.).*?(?=' + pattern1 + ')'
# pattern3 = r'(Willakuq:|Qillqaq:).*?(?=' + pattern1 + '|' + pattern2 + ')'

# Use re.sub to replace the matched text with an empty string
text = re.sub(pattern3, ' ', text, flags=re.DOTALL)
text = re.sub(pattern1, ' ', text)
text = re.sub(pattern2, '', text)
```

## Ayllupi yachasunchik

```python
# Define the patterns for the text to remove
# footer patterns
pattern1 = r'\x0c'
pattern2 = r'/\s*\w+'
pattern3 = r'\d?'

text = re.sub(pattern1, ' ', text)
text = re.sub(pattern2, '', text)
text = re.sub(pattern3, '', text)
```

```python
# remove the '-' in the text
text = re.sub(r'–|-|_', ' ', text) 
# make sure words are separated by a single space
text = re.sub(r'\s+', ' ', text)
# replace al the ’ with '
text = re.sub('’', "'", text)
# remove the the letters like 'a.'
text = re.sub(r'\b[a-z]\.', '', text)

text = text.strip()
text
```

# Tokenizacion

Filtrar oraciones en español

```python
from nltk.tokenize import sent_tokenize, word_tokenize
from langdetect import detect

sentences = sent_tokenize(text)
non_spanish_sentences = []

#Oraciones antes de filtrar
print("Oraciones antes de filtrar: ", len(sentences))
for sentence in sentences:
    try:
        if detect(sentence) != 'es':
            non_spanish_sentences.append(sentence)
    except:
        pass

sentences = non_spanish_sentences
print("Oraciones después de filtrar: ", len(sentences))
```

```python
def oraciones_mucho_espaniol(words):
    spanish_words = 0
    for word in words:
        try:
            if detect(word) == 'es':
                spanish_words += 1
        except:
            pass
    return spanish_words/len(words) > 0.5

def oraciones_muy_cortas(words, min_length=3):
    return len(words) > min_length

def oraciones_muy_repititivas(words, threshold=0.4):
    unique_words = set(words)
    ratio = len(unique_words) / len(words)
    return ratio >= threshold

def palabras_muy_largas(words, threshold=40):
    for word in words:
        if len(word) > threshold:
            return False
    return True

def split_tokens(sentence):
    # Check for three or more sequential tokens composed of one or two characters
    if re.search(r"(\b\w{1,2}\b\s){3,}", sentence):
        return False
    return True

def oraciones_con_matematica(sentence):
    if re.search(r"[\d+\-*/]+", sentence):
        return False
    return True
```

```python
def rule_based_heuristic(sentence):
    words = word_tokenize(sentence)
    
    valid = oraciones_muy_cortas(words) 
    valid = valid and oraciones_muy_repititivas(words)
    valid = valid and palabras_muy_largas(words)
    valid = valid and split_tokens(sentence)
    valid = valid and oraciones_con_matematica(sentence)

    return valid
```

```python
filtered_sentences = []

# filtros para las oraciones
for sentence in sentences:
    if rule_based_heuristic(sentence):
        filtered_sentences.append(sentence)

sentences = filtered_sentences

print("Oraciones después de filtrar: ", len(filtered_sentences))
```

```python
# eliminar oraciones repetidas
sentences = list(set(sentences))
print("Oraciones después de eliminar repetidas: ", len(sentences))
```

```python
# save the sentences to a file
with open('data/corpus/Ayllupi-yachasunchik-1-ñiqi-qullaw-qichwapi.txt', 'w', encoding='utf-8') as f:
    for sentence in sentences:
        f.write(sentence + '\n')
```
