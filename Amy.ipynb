```python
from pdfminer.high_level import extract_text, extract_pages
from pdfminer.layout import LAParams, LTTextBox, LTTextContainer
import re
```

```python
text_o = extract_text('data/Willakuykuna-mayt’u-Collao.pdf', page_numbers=range(8, 161))
text_o
```

```python
text_o = text_o.replace('\x0c', '\x0c\n')
lines = text_o.split('\n')
new_lines = [lines[0]] + [line for prev_line, line in zip(lines, lines[1:]) if line != prev_line]
text_o = '\n'.join(new_lines)
```

```python
# text_o to file
with open('data/Willakuykuna-mayt’u-Collao_2.txt', 'w', encoding='utf-8') as f:
    f.write(text_o)
```

# Limpieza de texto

```python
# get rid of the \n\n and \n in the text. make sure to put a ' ' in place of the \n\n 
text = re.sub(r'\n', ' ', text_o)
text
```

```python
# Define the patterns for the text to remove
# footer patterns
pattern1 = r'\d*?\s*\x0c'
pattern2 = r'/\s*\w+'
# author patterns
pattern3 = r'(Willakuq:|Qillqaq:|Willaquq:|Kastilla simipi qillqaq:|Qichwaman tikraspa qillqaq:|Chayllam chay.).*?(?=' + pattern1 + ')'
# pattern3 = r'(Willakuq:|Qillqaq:).*?(?=' + pattern1 + '|' + pattern2 + ')'

# Use re.sub to replace the matched text with an empty string
text = re.sub(pattern3, ' ', text, flags=re.DOTALL)
text = re.sub(pattern1, ' ', text)
text = re.sub(pattern2, '', text)
```

```python
# remove the '-' in the text
text = re.sub(r'–|-', ' ', text) 
# make sure words are separated by a single space
text = re.sub(r'\s+', ' ', text)
# replace al the ’ with '
text = re.sub('’', "'", text)

text = text.strip()
text
```

```python
# save the text to a file
with open('data/Willakuykuna-mayt’u-Collao_1.txt', 'w', encoding='utf-8') as f:
    f.write(text)
```

# Tokenizacion

Filtrar oraciones en español

```python
from nltk.tokenize import sent_tokenize, word_tokenize
from langdetect import detect

sentences = sent_tokenize(text)
non_spanish_sentences = []

#Oraciones antes de filtrar
print("Oraciones antes de filtrar: ", len(sentences))
for sentence in sentences:
    try:
        if detect(sentence) != 'es':
            non_spanish_sentences.append(sentence)
    except:
        pass

sentences = non_spanish_sentences
print("Oraciones después de filtrar: ", len(sentences))
```

```python
# save the sentences to a file
with open('data/corpus/Willakuykuna-mayt’u-Collao.txt', 'w', encoding='utf-8') as f:
    for sentence in sentences:
        f.write(sentence + '\n')
```

Basado en reglas

```python
alfabeto_quechua = ['a', 'aa', 'ch', 'chh', 'ch\'', 'ts', 'tr', 'h', 'i', 'ii', 'k', 'kh', 'k\'', 'l', 'll', 'm', 'n', 'ñ', 'p', 'ph', 'p\'', 'q', 'qh', 'q\'', 'r', 's', 'sh', 't', 'th', 't\'', 'u', 'uu', 'v', 'w', 'y']

def rule_based_heuristic(sentence):
    words = word_tokenize(sentence)
    #Eliminar oraciones con menos de 3 palabras
    if len(words) < 3:
        return False
    
    #Eliminar oraciones con palabras que no estén en el alfabeto quechua
    for word in words:
        for i, letter in enumerate(word):
            #Continue  if letter is not a letter
            if not letter.isalpha():
                continue
            if letter.lower() not in alfabeto_quechua:
                #Chequear siguiente letra
                if i+1 >= len(word):
                    return False
                letter = letter + word[i+1]
                if letter.lower() not in alfabeto_quechua:
                    if i+2 >= len(word):
                        return False
                    #Chequear siguiente letra
                    letter = letter + word[i+2]
                    if letter.lower() not in alfabeto_quechua:
                        return False
    return True
```

```python
# Filtrar oraciones con reglas heurísticas
filtered_sentences = []
for sentence in sentences:
    if rule_based_heuristic(sentence):
        filtered_sentences.append(sentence)
    else:
        print(sentence)

print("Oraciones después de filtrar con reglas heurísticas: ", len(filtered_sentences))
```
